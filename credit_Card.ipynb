```python
# Function to calculate Haversine distance
def haversine(lat1, lon1, lat2, lon2):
    # Convert latitude and longitude from degrees to radians
    lat1, lon1, lat2, lon2 = map(np.radians, [lat1, lon1, lat2, lon2])

    # Haversine formula
    dlat = lat2 - lat1
    dlon = lon2 - lon1
    a = np.sin(dlat / 2)**2 + np.cos(lat1) * np.cos(lat2) * np.sin(dlon / 2)**2
    c = 2 * np.arcsin(np.sqrt(a))
    r = 6371  # Radius of Earth in kilometers
    return c * r
```

```python
import numpy as np
# Step 3: Calculate distance from usual location
df2['distance_from_usual'] = df2.apply(
    lambda row: haversine(row['lat'], row['long'], row['usual_lat'], row['usual_lon']),
    axis=1
)
```

Get the user's usual transaction location: This can be the average or most frequent transaction latitude and longitude for the user.

Set a threshold: Define a maximum allowable distance (e.g., 50 km) to flag transactions as "unusual."
Mark suspicious transactions: If the distance exceeds the threshold, label the transaction as occurring far from the usual area.

#Autoencoders (Deep Learning)

```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix



# Assuming `X_balanced` and `y_balanced` are already prepared
scaler = StandardScaler()
X_balanced_scaled = scaler.fit_transform(X_balanced)

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_balanced_scaled, y_balanced, test_size=0.2, random_state=42)

# Extract only non-fraudulent transactions for training the autoencoder
X_train_normal = X_train[y_train == 0]


# Define input size based on the number of features
input_dim = X_train_normal.shape[1]

# Build the Autoencoder
input_layer = Input(shape=(input_dim,))
# Encoder
encoded = Dense(16, activation="relu")(input_layer)
encoded = Dense(8, activation="relu")(encoded)
# Decoder
decoded = Dense(16, activation="relu")(encoded)
decoded = Dense(input_dim, activation="sigmoid")(decoded)

# Model
autoencoder = Model(inputs=input_layer, outputs=decoded)
autoencoder.compile(optimizer="adam", loss="mse")

# Summary
autoencoder.summary()


# Train the autoencoder on normal transactions
history = autoencoder.fit(
    X_train_normal,
    X_train_normal,
    epochs=50,
    batch_size=256,
    validation_split=0.2,
    shuffle=True,
    verbose=1
)


# Calculate reconstruction errors
reconstructed_train = autoencoder.predict(X_train)
reconstruction_error_train = np.mean(np.power(X_train - reconstructed_train, 2), axis=1)

reconstructed_test = autoencoder.predict(X_test)
reconstruction_error_test = np.mean(np.power(X_test - reconstructed_test, 2), axis=1)

# Define a threshold for fraud detection
threshold = np.percentile(reconstruction_error_train, 95)  # 95th percentile
print(f"Reconstruction error threshold: {threshold}")

# Predict fraud based on reconstruction error
y_pred_test = (reconstruction_error_test > threshold).astype(int)



# Evaluation
print("Classification Report:")
print(classification_report(y_test, y_pred_test))

print("Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_test))

# Calculate ROC AUC Score
auc_score = roc_auc_score(y_test, reconstruction_error_test)
print(f"ROC AUC Score: {auc_score}")


```
